## Q1 (2pts)

The file global_temp.csv in Vocareum (obtained from https://github.com/datasets/global-temp) contains recordings of the average global temperature per year from 2 different sources.

Plot the series of temperatures as a line graph in ggplot, with measurements from each source a differently colored line. Set the width of the lines to 1.5 using the parameter 'size'. Use the following annotations for the plot:
+ Title of the plot: "Average Global Temperature in ºC per Year"
+ X axis: "Year"
+ Y axis: "Temperature (ºC)"


```python
###
### YOUR CODE HERE
###
df = pd.read_csv("global_temp.csv")
(ggplot() + geom_line(df,aes(x='Year', y='Mean',color='factor(Source)'), size=1.5)
+ ggtitle("Average Global Temperature in ºC per Year")
+ labs(x='Year', y='Temperature (ºC)'))
```

## Q2 (2+2+2+2)

Read in data from the file "housing.csv". 


1) Plot 1: Create a scatterplot showing Sqft on the x-axis, and Price on the y-axis. Use the following annotations for the plot:
+ Title of the plot: "Visualization of House prices"
+ X axis: "Area in square feet"
+ Y axis: "Price per square foot".

Set the size of the points to an appropriate value.

```python
###
### YOUR CODE HERE
###
df = pd.read_csv("housing.csv")
(ggplot() + geom_point(df,aes(x='Sqft',y='Price'))
+ ggtitle("Visualization of House prices")
+ labs(x='Area in square feet', y='Price per square foot'))
```

2) Plot 2: Repeat the above, now with the color and size of the circles controlled by Location and Baths respectively. Set the transparency of the circles to 0.6

```python
###
### YOUR CODE HERE
###
(ggplot() + geom_point(df,aes(x='Sqft',y='Price',color='factor(Location)',size='factor(Baths)'),alpha=0.6)
+ ggtitle("Visualization of House prices")
+ labs(x='Area in square feet', y='Price per square foot'))
```

3) Plot 3: Create a histogram showing the distribution of the values in the 'Price' column. Notice that the histogram (and the earlier scatterplot) indicates the data may have an outlier. Explore the impact of the number of bins on the ability to see the outlier. Print the minimum value of the number of bins when the outlier is first visible. Plot a histogram using this minimum number of bins.


```python
###
### YOUR CODE HERE
###
(ggplot() + geom_histogram(df,aes(x='Price'),bins=4, color='black',fill='orange'))
```

4) Plot 4: Remove the outlier (i.e., maximum value) from the 'Price' column and plot a boxplot of prices, one of each value of Location. Set the linewidth of the boxplots to 2, and fill them with the color gold.

```python
###
### YOUR CODE HERE
###
df_sub = df.drop(df.loc[df['Price']==df['Price'].max()].index)
(ggplot() + geom_boxplot(df_sub,aes(x='Location',y='Price'),fill='gold',size=2))
```

## Q3  (3+4)

Read in data from the file "cars_information.csv".

1) 
We want to look at the relation between hp, and mpg  for each setting of the column am (automatic or manual). Create a scatter plot of hp (x-axis) vs mpg (on the y-axis), each point colored differently according to the value of am. For each setting of am, also plot a straight line with error bars (using geom_smooth) (also colored differently according to am).

```python
###
### YOUR CODE HERE
###
df = pd.read_csv("cars_information.csv")
df

(
    ggplot(df, aes(x='hp', y='mpg',color='factor(am)'))
    + geom_point()
    + geom_smooth(method='lm')
)

```

2) Next, we want to simultaneously look at the relation between mpg (x-axis), and disp and hp (both on the y-axis). Create a plot with two facets, side by side in a single row. Both have mpg on the x-axis. The left facet has disp on the y-axis, and the right has hp on the y-axis. In each of these facets, we want the same two layers as before: a scatter plot and  a straight line using geom_smooth.

Hint: you will have to use the melt() method to concatenate the values in columns disp and hp into a single column. Melt should also create a second column indicating if that value is disp or hp. The value of the second column will tell ggplot which facet this belongs to (see the slides).

```python
###
### YOUR CODE HERE
###
# df
df_sub = df.melt(id_vars=['mpg'], value_vars=['disp','hp'])

(
    ggplot(df_sub, aes(x='mpg',y='value'))
    + geom_point()
    + geom_smooth(method='lm')
    + facet_grid('. ~ variable')
)
print(df_sub)
```

## Q4 (2+2+2+2+2)

1) Load the dataframe states.csv and print out it's dimensions. Also print out its first 10 rows. Columns long and lat have a series of latitudes and longitudes, plot these using the geom_path geometry (and NOT geom_line, though it is worth looking at what happens with this). Decide which axes lat and long should be mapped to.

```python
###
### YOUR CODE HERE
###

df = pd.read_csv('states.csv')

print(df.columns)
print(df.head(10))

(
    ggplot(df, aes(x='long',y='lat'))
    + geom_path()
)

# x-axis = long and y-axis = lat
```

2) To remove some of the crisscrossing lines, we will again use geom_path, now grouping by the column 'group'. You plot should show the state-boundaries perfectly. Set the opacity of the lines to 0.3. Store the result in us_plt. Also plot it.

```python
###
### YOUR CODE HERE
###
us_plt = (
    ggplot()
    + geom_path(df, aes(x='long',y='lat', group='group'),alpha=0.3)
)

print(us_plt)
```

3) Create a new dataframe state_center, with one row for each state, and with columns long and lat giving the mean latitude and longitude values for that state. Also, make sure that the state names are in a column called region.

```python
###
### YOUR CODE HERE
###

state_center = (df.groupby('region')
                .agg({'long':'mean','lat':'mean'}))

state_center
```
