## Part 2

For k = 1, 2, 3, 4, and 5 do the following: 
1. Perform k-means clustering. 
2. Create a scatter plot with feature 'V1' on x axis and feature 'V2' on y axis. For each scatter plot:
    i. Color each cluster with a different color. 
    ii. Mark the centroids obtained after running the k-means algorithm as red crosses.
    
Note: to get centroids, you must use the .cluster_centers attribute (see documentation). This will return a numpy array: use the first column as the x-coordinates of your centroids and second column as the y-coordinates.

```python
###
### YOUR CODE HERE
###

def km(k):
    kmeans = KMeans(k)
    rslt = kmeans.fit(features)
    c = rslt.predict(features)
    centers = kmeans.cluster_centers_
    center = pd.DataFrame(centers, columns=["col1", "col2"])

    plot = (ggplot() + geom_point(features,aes(x='V1',y='V2',color='c'))
               + geom_point(center,aes(x='col1',y='col2'),shape='x',size=3,color='red'))
    return plot

for i in range(1,6):
    print(km(i))
```
